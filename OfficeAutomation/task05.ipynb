- é—®é¢˜ï¼šå¦‚ä½•åœ¨ä¸¤ä¸ªå­—ç¬¦ä¸²"Tom123"å’Œ"Tom"ä¸­é€‰æ‹©å‰è€…ï¼Ÿ

````python
strings = ["Tom123", "Tom", "123Tom"]

print(f"To check for {strings[0]}:", re.match("Tom\d+", strings[0]))
print(f"To check for {strings[1]}:", re.match("Tom\d+", strings[1]))
print(f"To check for {strings[2]}:", re.match("Tom\d+", strings[2]))

print(f"To check for {strings[0]}:", re.match("Tom", strings[0]))
print(f"To check for {strings[1]}:", re.match("Tom", strings[1]))
print(f"To check for {strings[2]}:", re.match("Tom", strings[2]))

print(f"To search for {strings[0]}:", re.search("Tom", strings[0]))
print(f"To search for {strings[1]}:", re.search("Tom", strings[1]))
print(f"To search for {strings[2]}:", re.search("Tom", strings[2]))

print("Our choice is {}".format([item for item in strings if re.match("Tom\d+", item) is not None][0]))
````

### é€šå‘å®Œå–„ä¹‹è·¯ï¼šå°è£…æˆé€šç”¨çš„å‡½æ•°

å¦‚æœå¸Œæœ›çˆ¬è™«èƒ½å¤Ÿåº”ç”¨äºä¸åŒçš„ç½‘é¡µï¼Œé‚£ä¹ˆå¯ä»¥ä¿®æ”¹ä¸Šè¿°æ¡ˆä¾‹åˆ°ä¸€ä¸ªé€šç”¨çš„å‡½æ•°ã€‚æˆ‘ä»¬æ¥å°è¯•ä¸€ä¸‹ã€‚

````python
def get_web_page_table(website, table_css_selector, encoding="UTF-8", header=True):
    # to crawl the web page
    r = requests.get(website)

    # to correct coding
    r.encoding = encoding

    # to create BS object for parse HTML
    bs_obj = BeautifulSoup(r.text, 'lxml')
    
    # first step: find the table
    table = bs_obj.select(table_css_selector)[0]
    
    row_data = []
    for row in table.select("tr"):
        row_data.append([item.text for item in row.select("td")])
    
    row_data = []
    for row in table.select("tr"):
        # to add table header
        if len(row.select("th")) > 0:
            row_data.append([re.sub("\s", "", item.text) for item in row.select("th")])
        else:
            row_data.append([re.sub("\s", "", item.text) for item in row.select("td")])
            
    if header:
        return pd.DataFrame(row_data[1:], columns=row_data[0])
    else:
        return pd.DataFrame(row_data)
````
